```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv('veh0data.csv.csv')
df1 = pd.read_csv('veh1data.csv.csv')
df2 = pd.read_csv('veh2data.csv.csv')
df3 = pd.read_csv('veh3data.csv.csv')
df4 = pd.read_csv('veh4data.csv.csv')
```

```python
df0 = df0.drop_duplicates(subset=["max_co2", "time_st"], ignore_index=True)
df0 = df0.sort_values(by=['time_st'])

df1 = df1.drop_duplicates(subset=["max_co2", "time_st"], ignore_index=True)
df1 = df1.sort_values(by=['time_st'])

df2 = df2.drop_duplicates(subset=["max_co2", "time_st"], ignore_index=True)
df2 = df2.sort_values(by=['time_st'])

df3 = df3.drop_duplicates(subset=["max_co2", "time_st"], ignore_index=True)
df3 = df3.sort_values(by=['time_st'])

df4 = df4.drop_duplicates(subset=["max_co2", "time_st"], ignore_index=True)
df4 = df4.sort_values(by=['time_st'])



ax0 = df0.plot(x="time_st", y='max_co2', kind='line')	
ax1 = df1.plot(ax=ax0, x="time_st", y='max_co2', kind='line')	
ax2 = df2.plot(ax=ax1, x="time_st", y="max_co2")
ax3 = df2.plot(ax=ax2, x="time_st", y="max_co2")
ax4 = df3.plot(ax=ax3, x="time_st", y="max_co2")
ax3.legend(["Vehicle 1", "Vehicle 2", "Vehicle 3", "Vehicle 4", "Vehicle 5"])
```
